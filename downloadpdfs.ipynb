```python
import whitepaper_url as wu
import pandas as pd
import ast

df = pd.read_csv('./masterdf2.csv', index_col="name")
cryptoanalyse = wu.cryptoafresh()
```

```python
for i in range(len(df)):
    name = df.index.values[i]
    links = df['g_whitep'].apply(ast.literal_eval)[i]
    if links == []:
        continue
    k = 0
    for link in links: 
        if ".pdf" not in link:
            continue
        cryptoanalyse.getpdffromurl(link,name+" " + str(k))
        k+=1
```
